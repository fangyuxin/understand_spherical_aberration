```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sym.init_printing()
```

```python
class var():
    def __init__(self, str):
        self.s = sym.Symbol(str)
        self.v = None
```

首先，利用$EFL = 100$以及$c_1 = -c_2$联立求解出$c_1$和$c_2$的值。

```python
x = var('x')
K = var('K')
n = var('n')
d = var('d')
apt_d = var('AD')
mu = n
c1, c2 = var('c_1'), var('c_2')
R1, R2 = var('R_1'), var('R_2')
K1, K2 = var('K_1'), var('K_2')
delta, delta_P = var('delta'), var('\delta^{\prime}')

K.v = 1/100
n.v = 1.51680
d.v = 7.2
apt_d.v = 30
```

```python
def eq1(get_val=False): return c1.s + c2.s if get_val else c1.s + c2.s
def eq2(get_val=False): return K.v - (n.v - 1)*(c1.s - c2.s) if get_val \
else K.s - (n.s - 1)*(c1.s - c2.s)
```

```python
sym.solve([eq1(), eq2()], [c1.s, c2.s])
```

```python
c1.v, c2.v = sym.solve([eq1(True), eq2(True)], [c1.s, c2.s]).values()
```


下面求解主面的位置。利用公式:

$$\normalsize \delta^{\prime} = \frac{-n^{\prime}  d  K_1}{\mu  K}$$
以及
$$\normalsize \delta          = \frac{ n           d  K_2}{\mu  K}$$
求解可得：

```python
K1.v = c1.v*(n.v - 1)
K2.v = c2.v*(1-n.v)

delta_P.v = (-n.v*d.v*K1.v)/(mu.v*K.v)
delta.v   = ( n.v*d.v*K2.v)/(mu.v*K.v)
```

容易知道，两个主面相互重合。

以像方主面与光轴的交点$P^{\prime}$为坐标原点，即: $O = P^{\prime}$。并建立直角坐标系。下面求圆心$C_1$和$C_2$的位置。设透镜与光轴的两个交点从左到右依次为$Q，Q^{\prime}$，那么有: $QQ^{\prime} = d$。

圆心$C_1$的坐标为:
$$C_1(P^{\prime}C_1, 0)$$
其中，

$$
\begin{align}
P^{\prime}C_1  &= P^{\prime}Q^{\prime} + Q^{\prime}Q + QC_1^{\prime}\\
               &= -\delta^{\prime} - d + R_1\\
               &= R_1 -\delta^{\prime} - d
\end{align}
$$
      
圆心$C_2$的坐标为:
$$C_2(P^{\prime}C_2, 0)$$
其中，

$$
\begin{align}
P^{\prime}C_2  &= P^{\prime}Q^{\prime} + Q^{\prime}C_2^{\prime}\\
               &= -\delta^{\prime}+ R_2\\
               &= R_2 -\delta^{\prime}
\end{align}
$$

在这里，$R_2$小于零。

```python
C1, C2 = var('C_1'), var('C_2')
R1.v = 1/c1.v
R2.v = 1/c2.v
C1.v = R1.v - delta_P.v - d.v
C2.v = R2.v - delta_P.v
```

```python
ang = np.linspace(0, 2*np.pi, 500)
plt.figure(figsize=(19/2, 5), dpi=200)
plt.plot(C1.v+ R1.v*np.sin(ang), R1.v*np.cos(ang))
plt.plot(C2.v+ R2.v*np.sin(ang), R2.v*np.cos(ang))
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'  ].set_position(('data',0))
plt.show()
```

```python
theta = var('theta')
def eq1(get_sym=True): return R1.s*sym.sin(theta.s) - apt_d.s/2 if get_sym \
    else R1.v*sym.sin(theta.s) - apt_d.v/2

def eq2(get_sym=True): return R1.s*sym.sin(theta.s) + apt_d.s/2 if get_sym \
    else R1.v*sym.sin(theta.s) + apt_d.v/2
```

```python
sym.solve(eq1(), theta.s)
```

```python
sym.solve(eq2(), theta.s)
```

```python
sym.solve(eq1(False), theta.s)
```

```python
_, theta_lower = sym.solve(eq1(False), theta.s)
_, theta_upper = sym.solve(eq2(False), theta.s)
theta_lower = float('%0.10f'%theta_lower)
theta_upper = float('%0.10f'%theta_upper)
```

```python
ang = np.arange(theta_lower, theta_upper, 0.001)
xx = np.arange(C1.v+ R1.v*np.cos(theta_lower), C2.v+ R2.v*np.cos(theta_lower), 0.001)

plt.figure(figsize=(5, 5), dpi=500)
plt.plot(C1.v+ R1.v*np.cos(ang), R1.v*np.sin(ang), 'b')
plt.plot(C2.v+ R2.v*np.cos(ang), R2.v*np.sin(ang), 'b')
plt.plot(xx, apt_d.v/2 + 0*xx, 'b')
plt.plot(xx, -apt_d.v/2 + 0*xx, 'b')

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'  ].set_position(('data',0))

plt.xlim((-20, 20))
plt.ylim((-40, 40))
plt.show()
```

```python
ang = np.arange(theta_lower, theta_upper, 0.001)
l = np.arange(C1.v+ R1.v*np.cos(theta_lower), C2.v+ R2.v*np.cos(theta_lower), 0.001)
xx0 = np.arange(-30, 30, 0.001)
xx1 = np.arange(-30, C1.v+ R1.v*np.cos(theta_lower), 0.001)

plt.figure(figsize=(5, 5), dpi=500)
plt.plot(C1.v + R1.v*np.cos(ang), R1.v*np.sin(ang), 'b', linewidth=0.8)
plt.plot(C2.v + R2.v*np.cos(ang), R2.v*np.sin(ang), 'b', linewidth=0.8)
plt.plot(l, apt_d.v/2 + 0*l, 'b', linewidth=0.8)
plt.plot(l, -apt_d.v/2 + 0*l, 'b', linewidth=0.8)
plt.plot(xx0, slope_1*(xx0 - x0.v) + apt_d.v/2, linewidth=0.8)

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'  ].set_position(('data',0))

plt.xlim((-20, 20))
plt.ylim((-40, 40))
plt.show()
```

由几何关系知，入射角$i_{1}$可以通过如下公式计算：
$$
sini_{1} = \frac{1}{\sqrt{1 + slope_{1}^{2}}}
$$
而$slope_{1}$就是图中切线的斜率，刚刚已经求得。


由折射定律可求得第一次折射的折射角$i^{'}_{1}$:
$$
sini_{1} = nsini^{'}_{1}
$$

```python
sini1 = 1/np.sqrt(float('%0.10f'%(1 + slope_1**2)))
```

```python
i1_P = np.arcsin(sini1/n.v)
i1   = np.arcsin(sini1)
```

```python
theta_1 = i1_P - i1
```

```python
slope_2 = np.tan(theta_1)
```

```python
ang = np.arange(theta_lower, theta_upper, 0.001)
l = np.arange(C1.v+ R1.v*np.cos(theta_lower), C2.v+ R2.v*np.cos(theta_lower), 0.001)
xx0 = np.arange(-30, 30, 0.001)
xx1 = np.arange(-30, C1.v+ R1.v*np.cos(theta_lower), 0.001)

plt.figure(figsize=(5, 5), dpi=100)
plt.plot(C1.v+ R1.v*np.cos(ang), R1.v*np.sin(ang), 'b', linewidth=0.8)
plt.plot(C2.v+ R2.v*np.cos(ang), R2.v*np.sin(ang), 'b', linewidth=0.8)
plt.plot(l, apt_d.v/2 + 0*l, 'b', linewidth=0.8)
plt.plot(l, -apt_d.v/2 + 0*l, 'b', linewidth=0.8)
plt.plot(xx0, slope_1*(xx0 - x0.v) + apt_d.v/2, 'g--', linewidth=0.5)
plt.plot(xx0, (-1/slope_1)*(xx0 - x0.v) + apt_d.v/2, 'g--', linewidth=0.5)
plt.plot(xx1,  apt_d.v/2 + 0*xx1, 'r', linewidth=0.8)

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'  ].set_position(('data',0))

plt.xlim((-20, 20))
plt.ylim((-20, 20))
plt.show()
```
